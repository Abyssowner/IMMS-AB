```python
# 计算 mae mse pearson r2 
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr
import math

# 读取CSV文件
try:
    df = pd.read_csv('dataset_paddle.csv')
    print(f"成功读取文件，共 {len(df)} 行")
except Exception as e:
    print(f"读取文件时出错: {str(e)}")
    exit(1)

# 显示前几行数据，了解数据结构
print("\n数据预览:")
print(df.head())

# 确保至少有两行数据（跳过第一行）
if len(df) < 2:
    print("数据行数不足，无法计算")
    exit(1)

# 从第二行开始提取第五列和第六列
column5 = df.iloc[1:, 4].values  # 第五列（索引为4）
column6 = df.iloc[1:, 5].values  # 第六列（索引为5）

# 打印列名，帮助理解数据
print(f"\n第五列名称: {df.columns[4]}")
print(f"第六列名称: {df.columns[5]}")

# 移除可能的NaN值
valid_indices = ~(np.isnan(column5) | np.isnan(column6))
column5 = column5[valid_indices]
column6 = column6[valid_indices]

# 检查是否还有数据
if len(column5) == 0 or len(column6) == 0:
    print("过滤NaN后没有有效数据")
    exit(1)

# 第五列作为真实值，第六列作为预测值，计算结果如下
print("\n反转后的计算结果 (以第五列为真实值):")
mae_rev = mean_absolute_error(column5, column6)
mse_rev = mean_squared_error(column5, column6)
rmse_rev = math.sqrt(mse_rev)
pearson_corr, _ = pearsonr(column5, column6)
r2_rev = r2_score(column5, column6)

print(f"MAE (平均绝对误差): {mae_rev:.6f}")
print(f"MSE (均方误差): {mse_rev:.6f}")
print(f"RMSE (均方根误差): {rmse_rev:.6f}")
print(f"Pearson相关系数: {pearson_corr:.6f}")
print(f"R² (决定系数): {r2_rev:.6f}")
```
